```python
test_size_num = 3827
all = 7653
def Z(classif):
  tn1 = 0
  fn1 = 0
  fp1 = 0
  tp1 = 0
  percent1 = 0
  f11 = 0
  for seed in seeds:
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.5, random_state=seed)
    classif.fit(X_train, y_train)
    y_pred = classif.predict(X_test)
    matrix = confusion_matrix(y_test, y_pred)
    
    tn1 += (matrix[0][0])
    fn1 += (matrix[0][1])
    fp1 += (matrix[1][0])
    tp1 += (matrix[1][1])
    percent1 += (1 - (matrix[0][1] + matrix[1][0])/test_size_num)

    prec = matrix[1][1]/(matrix[1][1] + matrix[1][0])
    recl = matrix[1][1]/(matrix[1][1] + matrix[0][1])
    f11 += (2*prec*recl/(prec+recl))


    
    X_test, X_train, y_test, y_train = train_test_split(X, Y, test_size = 0.5, random_state=seed)
    classif.fit(X_train, y_train)
    y_pred = classif.predict(X_test)
    matrix = confusion_matrix(y_test, y_pred)
    
    tn1 += (matrix[0][0])
    fn1 += (matrix[0][1])
    fp1 += (matrix[1][0])
    tp1 += (matrix[1][1])
    percent1 += (1 - (matrix[0][1] + matrix[1][0])/test_size_num)

    prec = matrix[1][1]/(matrix[1][1] + matrix[1][0])
    recl = matrix[1][1]/(matrix[1][1] + matrix[0][1])
    f11 += (2*prec*recl/(prec+recl))


  
  tn.append(tn1/200)
  fn.append(fn1/200)
  fp.append(fp1/200)
  tp.append(tp1/200)
  f1.append(f11/200)
  percent.append(percent1/200)
```
